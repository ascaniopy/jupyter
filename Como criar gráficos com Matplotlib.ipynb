```python
# Preencher dados da coluna Medal com a string "Nenhuma".
dados['Medal'].fillna('Nenhuma', inplace=True)
# Preencher dados da coluna Age com a m√©dia das idades.
dados['Age'].fillna(dados['Age'].mean(), inplace=True)
# Preencher dados da coluna Height com a m√©dia das alturas.
dados['Height'].fillna(dados['Height'].mean(), inplace=True)
# Preencher dados da coluna Weight com a m√©dia dos pesos.
dados['Weight'].fillna(dados['Weight'].mean(), inplace=True)
```

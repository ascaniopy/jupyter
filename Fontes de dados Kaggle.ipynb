```python
import pandas as pd
```

```python
cotacao_df = pd.read_csv('indexData.csv')
display(cotacao_df)
```

```python
cotacao_df1 = cotacao_df[['Index', 'Date', 'Open', 'Close']]
display(cotacao_df1)
```

```python
dfremove = cotacao_df1.loc[(cotacao_df1['Index'] != 'NYA')] # Remove todas as linhas diferentes de "NYA".
cotacao_dffinal = cotacao_df1.drop(dfremove.index)   # Mostra o novo DataFrame.
display(cotacao_dffinal) # Mostra só as linhas com "NYA".
cotacao_dffinal = cotacao_dffinal[13900:]  # Mostra somente as últimas 48 linhas.
display(cotacao_dffinal)  # Mostra o novo DataFrame.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 4))
plt.plot(cotacao_dffinal['Date'], cotacao_dffinal['Close'], label='NYA', color='g', ls='--', lw='2')
plt.legend(loc=2, fontsize=18)
plt.ylabel('Valor de cotação', fontsize=14)
plt.xlabel('Data da cotação', fontsize=14)
plt.title('Gráfico de cotação histórica da Bolsa NYA', fontsize=18)
plt.xticks(['2021-03-23', '2021-04-23', '2021-05-23'])
plt.show()
```

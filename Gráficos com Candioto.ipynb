```python
import matplotlib.pyplot as plt
```

```python
# Gráfico de linha.
x = [1, 2, 3, 4]
y = [2, 3, 4, 3]
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 3]
plt.plot(x, y, label ='dados')
plt.ylabel('Eixo y')
plt.xlabel('Eixo x')
plt.title('Título do Gráfico')
plt.xticks([0, 2, 4, 6, 8, 9])  # Aqui agora é meu Eixo X.
plt.yticks([-1, 3, 5, 9, 11])   # Aqui agora é meu Eixo Y.
plt.legend()
plt.show()
```

```python
plt.plot(x, y, linestyle='solid', color='green')
plt.show()
```

```python
plt.plot(x, y, linestyle='dashed', color='red')
plt.show()
```

```python
plt.plot(x, y, linestyle=':', color='black', lw = 5.0)
plt.show()
```

```python
# Gráfico de dispersão.
plt.scatter(x, y, color='blue')
plt.show()
```

```python
# Gráfico de barras.
plt.bar(x, y, color='cyan')
plt.show()
```

Alterando os limites dos gráficos.


```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 3]
plt.plot(x, y, color='red')
plt.axis(xmin=-1, xmax=10, ymin=0, ymax=12)
plt.show()
```

Subplots

```python
valoresx = [1, 2, 3, 4]
valoresy = [1, 4, 2,3]

figura = plt.figure(figsize=(20, 5))
figura.suptitle('Título Geral')

figura.add_subplot(131) #  1: Uma linha, 3: três colunas, 1: posição um.
plt.plot(valoresx, valoresy, label='Dado em linha.')
plt.legend()
plt.title('Gráfico 1')

figura.add_subplot(132) #  1: Uma linha, 3: três colunas, 1: posição dois.
plt.bar(valoresx, valoresy, label='Dado em colulas.')
plt.legend()
plt.title('Gráfico 2')

figura.add_subplot(133) #  1: Uma linha, 3: três colunas, 1: posição três.
plt.scatter(valoresx, valoresy, label='Dados em pontos.')
plt.legend()
plt.title('Gráfico 3')

plt.savefig('raficos.png') # CTRL + ENTER para salvar, vai salvar na mesma pasta.

plt.show(figura)
```
